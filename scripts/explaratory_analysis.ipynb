```python
from google.colab import drive
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
drive.mount('/content/drive')
file_path = '/content/drive/My Drive/arxiv-rag-project/cs_papers_50k_with_institutions_final1.json'
df = pd.read_json(file_path)
print(df.head())
```

```python
region_counts = df['region'].value_counts()
print(region_counts)
```

```python
top_institutions = df['primary_institution'].value_counts().head(20)
print(top_institutions)
```

```python
region_counts.sort_values().plot(
    kind='barh',
    figsize=(10,6)
)

plt.xlabel("Number of Papers")
plt.title("Region Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,8))

bars = plt.barh(
    top_institutions.sort_values().index,
    top_institutions.sort_values().values
)

plt.xlabel("Number of Papers")
plt.title("Top 20 Institutions by Publication Count")

plt.gca().spines[['top','right']].set_visible(False)

plt.tight_layout()
plt.show()
```
