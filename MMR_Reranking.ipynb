```python
"""
CELL 2 — Updated for CS neutral queries
Changes vs previous Cell 2:
  1. QUERY_FILE  → cs_neutral_queries.json
  2. OUTPUT_DIR  → data2/cs_results  (keeps old results safe)
  3. NDCG@10 and MRR added to metrics
"""

# STEP 1 - MOUNT GOOGLE DRIVE
from google.colab import drive
drive.mount("/content/drive")
print(" Google Drive mounted")

# STEP 2 - IMPORTS
import os, json
import numpy as np
import pandas as pd
from itertools import combinations
from tqdm import tqdm
from sentence_transformers import SentenceTransformer
from llama_index.core import VectorStoreIndex, StorageContext, Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.vector_stores.chroma import ChromaVectorStore
import chromadb
print(" All imports successful")

# STEP 3 - PATHS & CONFIG
CHROMA_PATH     = "/content/drive/MyDrive/arxiv-rag-project/data2/chroma_db"
QUERY_FILE      = "/content/drive/MyDrive/arxiv-rag-project/data2/cs_neutral_queries.json"  # ← CHANGED
OUTPUT_DIR      = "/content/drive/MyDrive/arxiv-rag-project/data2/cs_results"               # ← CHANGED
COLLECTION_NAME = "cs_papers_v4"
EMBED_MODEL     = "all-MiniLM-L6-v2"
N_QUERIES       = 100
TOP_K_FETCH     = 50
TOP_K_MMR       = 10
LAMBDAS         = [0.3, 0.5, 0.7, 0.9]
CORPUS_BASELINE = 0.58

os.makedirs(OUTPUT_DIR, exist_ok=True)
print(f"✓ Config ready | Output → {OUTPUT_DIR}")

# STEP 4 - LOAD EMBEDDING MODEL
print("\nLoading embedding model …")
st_model = SentenceTransformer(EMBED_MODEL)
print(f" '{EMBED_MODEL}' loaded")

# STEP 5 - LLAMAINDEX + CHROMADB SETUP
print("\nConnecting to ChromaDB …")
Settings.embed_model = HuggingFaceEmbedding(model_name=EMBED_MODEL)
Settings.llm         = None

chroma_client     = chromadb.PersistentClient(path=CHROMA_PATH)
chroma_collection = chroma_client.get_collection(COLLECTION_NAME)
print(f" Collection '{COLLECTION_NAME}' → {chroma_collection.count()} docs")

vector_store    = ChromaVectorStore(chroma_collection=chroma_collection)
storage_context = StorageContext.from_defaults(vector_store=vector_store)
index           = VectorStoreIndex.from_vector_store(
                      vector_store, storage_context=storage_context)
retriever_50 = index.as_retriever(similarity_top_k=TOP_K_FETCH)
print(f" Retriever ready (top_k={TOP_K_FETCH})")

# STEP 6 - LOAD QUERIES
print(f"\nLoading queries …")
with open(QUERY_FILE, "r", encoding="utf-8") as f:
    all_queries = json.load(f)
queries = all_queries[:N_QUERIES]
print(f" {len(queries)} queries loaded")
print(f"  Categories: { dict(pd.Series([q['primary_cat'] for q in queries]).value_counts()) }")


# ══════════════════════════════════════════════════════════════════
# HELPER FUNCTIONS
# ══════════════════════════════════════════════════════════════════

def cosine_sim(a: np.ndarray, b: np.ndarray) -> float:
    """Cosine similarity between two normalised vectors."""
    denom = np.linalg.norm(a) * np.linalg.norm(b)
    return float(np.dot(a, b) / denom) if denom > 1e-10 else 0.0


def mmr_rerank(query_emb, doc_embs, lam, k):
    """
    Maximal Marginal Relevance (Carbonell & Goldstein, 1998).
    MMR(d_i) = λ·sim(d_i,query) − (1−λ)·max_{d_j∈S} sim(d_i,d_j)
    Returns list of k indices into doc_embs.
    """
    n         = len(doc_embs)
    rel       = np.array([cosine_sim(query_emb, doc_embs[i]) for i in range(n)])
    selected  = []
    remaining = list(range(n))
    for _ in range(min(k, n)):
        best_idx, best_score = None, -np.inf
        for idx in remaining:
            red   = max((cosine_sim(doc_embs[idx], doc_embs[s]) for s in selected), default=0.0)
            score = lam * rel[idx] - (1 - lam) * red
            if score > best_score:
                best_score, best_idx = score, idx
        selected.append(best_idx)
        remaining.remove(best_idx)
    return selected


def compute_mmr_metrics(query_emb, selected_embs, lam):
    """
    avg_relevance  : mean cosine-sim(doc, query)
    intra_list_div : mean pairwise cosine-distance
    mmr_score      : λ·avg_relevance + (1−λ)·ILD
    """
    k       = len(selected_embs)
    avg_rel = float(np.mean([cosine_sim(query_emb, e) for e in selected_embs]))
    ild     = (float(np.mean([1.0 - cosine_sim(selected_embs[i], selected_embs[j])
                               for i, j in combinations(range(k), 2)]))
               if k >= 2 else 0.0)
    return {
        "avg_relevance"  : round(avg_rel, 6),
        "intra_list_div" : round(ild, 6),
        "mmr_score"      : round(lam * avg_rel + (1 - lam) * ild, 6),
    }


def compute_ndcg_at_k(ranked_embs, query_emb, k=10, threshold=0.5):
    """
    NDCG@K using cosine similarity as relevance proxy.

    Since we don't have binary ground truth here (that's computed
    in Cell 3 via GPT), we use cosine-sim as a graded relevance score.

    rel_i = cosine_sim(doc_i, query)   (already in [0,1])
    DCG   = Σ rel_i / log2(i+2)        for i in 0..k-1
    IDCG  = DCG of perfect ranking (sort rel scores descending)
    NDCG  = DCG / IDCG
    """
    k      = min(k, len(ranked_embs))
    rels   = np.array([cosine_sim(query_emb, ranked_embs[i]) for i in range(k)])
    dcg    = float(np.sum(rels / np.log2(np.arange(2, k + 2))))
    # ideal = sort descending
    ideal  = np.sort(rels)[::-1]
    idcg   = float(np.sum(ideal / np.log2(np.arange(2, k + 2))))
    return round(dcg / idcg, 6) if idcg > 0 else 0.0


def compute_mrr(ranked_embs, query_emb, threshold=0.5):
    """
    MRR (Mean Reciprocal Rank).

    Finds the rank of the FIRST doc with cosine_sim >= threshold.
    MRR = 1 / rank_of_first_relevant_doc
    If no relevant doc found → MRR = 0

    threshold=0.5 is a reasonable cutoff for cosine similarity
    to determine relevance without GPT calls.
    """
    for i, emb in enumerate(ranked_embs):
        if cosine_sim(query_emb, emb) >= threshold:
            return round(1.0 / (i + 1), 6)
    return 0.0


# ══════════════════════════════════════════════════════════════════
# MAIN LOOP
# ══════════════════════════════════════════════════════════════════
print(f"\nRunning MMR reranking …")
print(f"  Queries={N_QUERIES} | Fetch={TOP_K_FETCH} | MMR k={TOP_K_MMR} | λ={LAMBDAS}\n")

rows    = []
skipped = 0

for q in tqdm(queries, desc="Queries"):
    qid, qtext = q["query_id"], q["query"]

    try:
        nodes = retriever_50.retrieve(qtext)
    except Exception as e:
        print(f"\n  [WARN] {qid}: {e}")
        skipped += 1
        continue
    if not nodes:
        skipped += 1
        continue

    node_texts = [n.get_content()[:500] for n in nodes]
    node_privs = [n.node.metadata.get("privilege", "Unknown") for n in nodes]

    cand_embs = st_model.encode(
        node_texts, normalize_embeddings=True,
        show_progress_bar=False, batch_size=64)
    q_emb = st_model.encode(qtext, normalize_embeddings=True)

    # baseline (top-10 cosine, no MMR)
    base_embs       = cand_embs[:TOP_K_MMR]
    base_metrics    = compute_mmr_metrics(q_emb, base_embs, lam=1.0)
    base_ndcg       = compute_ndcg_at_k(base_embs, q_emb, k=TOP_K_MMR)
    base_mrr        = compute_mrr(base_embs, q_emb)
    base_priv_cnt   = sum(1 for p in node_privs[:TOP_K_MMR] if p == "Privileged")

    for lam in LAMBDAS:
        sel_idx  = mmr_rerank(q_emb, cand_embs, lam=lam, k=TOP_K_MMR)
        sel_embs = cand_embs[sel_idx]
        metrics  = compute_mmr_metrics(q_emb, sel_embs, lam=lam)

        # NDCG@10 on MMR-selected docs
        ndcg = compute_ndcg_at_k(sel_embs, q_emb, k=TOP_K_MMR)

        # MRR on MMR-selected docs
        mrr  = compute_mrr(sel_embs, q_emb)

        sel_privs = [node_privs[i] for i in sel_idx]
        priv_cnt  = sum(1 for p in sel_privs if p == "Privileged")

        rows.append({
            "query_id"             : qid,
            "query_type"           : q.get("query_type", ""),
            "primary_cat"          : q.get("primary_cat", ""),
            "region"               : q.get("region", ""),
            "query_privilege"      : q.get("privilege", ""),
            "lambda"               : lam,
            "n_candidates"         : len(nodes),
            # ── MMR embedding metrics ─────────────────────────
            "avg_relevance"        : metrics["avg_relevance"],
            "intra_list_div"       : metrics["intra_list_div"],
            "mmr_score"            : metrics["mmr_score"],
            # ── NDCG & MRR (NEW) ──────────────────────────────
            "ndcg_at_10"           : ndcg,
            "mrr"                  : mrr,
            # ── Baseline comparison ───────────────────────────
            "base_avg_rel"         : base_metrics["avg_relevance"],
            "base_ild"             : base_metrics["intra_list_div"],
            "base_ndcg"            : base_ndcg,
            "base_mrr"             : base_mrr,
            # ── Privilege distribution ─────────────────────────
            "mmr_privileged_count" : priv_cnt,
            "mmr_privileged_pct"   : round(priv_cnt / TOP_K_MMR * 100, 1),
            "base_privileged_count": base_priv_cnt,
            "base_privileged_pct"  : round(base_priv_cnt / TOP_K_MMR * 100, 1),
        })

print(f"\n {N_QUERIES-skipped}/{N_QUERIES} queries done  ({skipped} skipped)")

# ── Save CSV ───────────────────────────────────────────────────────
df = pd.DataFrame(rows)
df.to_csv(os.path.join(OUTPUT_DIR, "mmr_results_per_query.csv"), index=False)

# ── Aggregate ──────────────────────────────────────────────────────
MCOLS = ["avg_relevance","intra_list_div","mmr_score",
         "ndcg_at_10","mrr",
         "base_avg_rel","base_ild","base_ndcg","base_mrr",
         "mmr_privileged_pct","base_privileged_pct"]

agg1 = df.groupby("lambda")[MCOLS].mean().round(6).reset_index()
agg1.to_csv(os.path.join(OUTPUT_DIR, "mmr_lambda_comparison.csv"), index=False)

df.groupby(["primary_cat","lambda"])[["avg_relevance","intra_list_div","ndcg_at_10","mrr"]]\
  .mean().round(6).reset_index()\
  .to_csv(os.path.join(OUTPUT_DIR, "mmr_by_category.csv"), index=False)

df.groupby(["query_privilege","lambda"])[["avg_relevance","ndcg_at_10","mrr","mmr_privileged_pct"]]\
  .mean().round(6).reset_index()\
  .to_csv(os.path.join(OUTPUT_DIR, "mmr_by_privilege.csv"), index=False)

# ── Print table ────────────────────────────────────────────────────
W = 10
print("\n" + "="*75)
print("  RESULTS: MMR metrics per Lambda  (CS neutral queries)")
print("="*75)
print(f"  {'λ':>5} | {'AvgRel':>{W}} | {'ILD':>{W}} | {'NDCG@10':>{W}} | {'MRR':>{W}} | {'Priv%':>{W}}")
print("  " + "─"*65)
for _, r in agg1.iterrows():
    print(f"  {r['lambda']:>5} | {r['avg_relevance']:>{W}.6f} | "
          f"{r['intra_list_div']:>{W}.6f} | {r['ndcg_at_10']:>{W}.6f} | "
          f"{r['mrr']:>{W}.6f} | {r['mmr_privileged_pct']:>{W}.1f}")

print(f"\n  Baseline → NDCG@10={agg1['base_ndcg'].mean():.4f}  MRR={agg1['base_mrr'].mean():.4f}")
print("="*75)
print(f"\n Cell 2 done! Output → {OUTPUT_DIR}")
```

```python
"""
CELL 3 — P@10 + SRR + SPD + NDCG@10 + MRR  for all lambda values
══════════════════════════════════════════════════════════════════
Self-contained — all functions defined here.
Needs from Cell 2 in memory:
  ✓ retriever_50, st_model, queries
  ✓ mmr_rerank(), cosine_sim()
  ✓ TOP_K_MMR, LAMBDAS, CORPUS_BASELINE, OUTPUT_DIR
"""

!pip install -q llama-index-llms-openai

import os, json, re, unicodedata
import numpy as np
import pandas as pd
from tqdm import tqdm
from llama_index.core.evaluation import RelevancyEvaluator
from llama_index.llms.openai import OpenAI

# ── Privilege label function (self-contained) ─────────────────────
TOP_100_CS_UNIVERSITIES = [
    "massachusetts institute of technology","stanford university","carnegie mellon university",
    "university of california berkeley","harvard university","princeton university",
    "university of california los angeles","university of washington","cornell university",
    "georgia institute of technology","university of texas austin","university of michigan ann arbor",
    "university of illinois urbana champaign","columbia university","university of california san diego",
    "yale university","university of pennsylvania","california institute of technology",
    "university of southern california","duke university","northwestern university","new york university",
    "university of wisconsin madison","university of maryland college park","purdue university",
    "university of massachusetts amherst","brown university","rice university","university of virginia",
    "ohio state university","penn state university","university of north carolina chapel hill",
    "university of california irvine","university of california santa barbara","university of california davis",
    "rutgers university","university of minnesota","texas a&m university","university of florida",
    "university of colorado boulder","boston university","university of rochester",
    "university of cambridge","university of oxford","imperial college london","ucl",
    "university college london","university of edinburgh","king's college london","university of glasgow",
    "university of sheffield","university of manchester","university of warwick","university of southampton",
    "university of bristol","eth zurich","kth royal institute of technology","university of helsinki",
    "epfl","technical university of munich","university of amsterdam","delft university of technology",
    "katholieke universiteit leuven","sorbonne university","university of copenhagen","karolinska institute",
    "lund university","university of zurich","rwth aachen","technische universitat berlin",
    "sapienza university of rome","politecnico di milano","university of barcelona",
    "national university of singapore","university of science and technology of china","tsinghua university",
    "peking university","university of tokyo","kyoto university","seoul national university","kaist",
    "fudan university","shanghai jiao tong university","zhejiang university","chinese university of hong kong",
    "university of hong kong","nanyang technological university","hong kong university of science and technology",
    "pohang university of science and technology","tokyo institute of technology","osaka university",
    "tohoku university","indian institute of technology ahmedabad","indian institute of technology chennai",
    "indian institute of technology kolkata","indian institute of technology bombay",
    "indian institute of technology kanpur","indian institute of technology kharagpur",
    "indian institute of technology delhi","indian institute of science","university of michigan-ann arbor",
    "sandia national laboratories","university of toronto","university of montreal",
    "university of british columbia","mcgill university","university of alberta","university of waterloo",
    "australian national university","university of melbourne","university of sydney",
    "university of new south wales","university of queensland","monash university",
    "university of california, san diego","technion","hebrew university of jerusalem",
    "tel aviv university","weizmann institute","google","meta","microsoft","amazon","deepmind",
    "openai","apple","nvidia","ibm","intel","max planck","allen institute","broad institute",
    "salk institute","scripps research"
]
_ALIASES  = {
    "mit":"massachusetts institute of technology","stanford":"stanford university",
    "berkeley":"university of california berkeley","cmu":"carnegie mellon university",
    "caltech":"california institute of technology","gatech":"georgia institute of technology",
    "uiuc":"university of illinois urbana champaign","umich":"university of michigan ann arbor",
    "uw":"university of washington","ucsd":"university of california san diego",
    "ucla":"university of california los angeles","usc":"university of southern california",
    "eth":"eth zurich","nus":"national university of singapore","ntu":"nanyang technological university",
    "hkust":"hong kong university of science and technology","cuhk":"chinese university of hong kong"
}
_STOPWORDS = {"university","of","the","institute","technology","college","school"}

def _normalize(text):
    if not text: return ""
    text = unicodedata.normalize('NFKD', text).lower()
    text = re.sub(r'\(.*?\)', '', text)
    text = re.sub(r'[^a-z0-9 ]', ' ', text)
    return re.sub(r'\s+', ' ', text).strip()

_TOP_SET      = set(_normalize(u) for u in TOP_100_CS_UNIVERSITIES)
_ALIASES_NORM = {_normalize(k): _normalize(v) for k, v in _ALIASES.items()}
_TOP_ORGS     = set(_normalize(x) for x in [
    "google","meta","microsoft","amazon","openai","deepmind","nvidia","apple","intel","ibm",
    "max planck institute","broad institute","allen institute","weizmann institute",
    "sandia national laboratories","lawrence berkeley national laboratory","argonne national laboratory"
])

def get_privilege_label(inst):
    if not inst or inst == "Unknown": return "Underrepresented"
    c = _normalize(inst)
    if c in _ALIASES_NORM: c = _ALIASES_NORM[c]
    if c in _TOP_SET: return "Privileged"
    for org in _TOP_ORGS:
        if org in c: return "Privileged"
    for top in _TOP_SET:
        if len((set(c.split()) - _STOPWORDS) & (set(top.split()) - _STOPWORDS)) >= 1:
            return "Privileged"
    return "Underrepresented"


# ── OpenAI key + evaluator ─────────────────────────────────────────
os.environ["OPENAI_API_KEY"] = "your-openai-api-key-here"

print("Setting up RelevancyEvaluator (GPT-4o-mini) …")
evaluator = RelevancyEvaluator(llm=OpenAI(model="gpt-4o-mini", temperature=0.3))
print(" Evaluator ready\n")


# ══════════════════════════════════════════════════════════════════
# METRIC HELPER FUNCTIONS
# ══════════════════════════════════════════════════════════════════

def compute_precision_at_k(query, nodes, k=TOP_K_MMR):
    """
    P@K using GPT-4o-mini RelevancyEvaluator.
    grade = 1 if result.passing else 0
    P@K = sum(grades) / K
    """
    grades = []
    for node in nodes[:k]:
        content = node.get_content()[:500]
        try:
            result = evaluator.evaluate(query=query, response=content, contexts=[content])
            grades.append(1 if result.passing else 0)
        except Exception as e:
            print(f"    [WARN] {e} — marking 0")
            grades.append(0)
    return round(sum(grades) / k, 6), grades


def compute_ndcg_at_k_gpt(grades, k=TOP_K_MMR):
    """
    NDCG@K using binary GPT grades (0 or 1).

    DCG  = Σ grade_i / log2(i+2)   for i in 0..k-1
    IDCG = DCG of perfect ranking (all 1s first)
    NDCG = DCG / IDCG
    """
    k      = min(k, len(grades))
    rels   = np.array(grades[:k], dtype=float)
    dcg    = float(np.sum(rels / np.log2(np.arange(2, k + 2))))
    ideal  = np.sort(rels)[::-1]
    idcg   = float(np.sum(ideal / np.log2(np.arange(2, k + 2))))
    return round(dcg / idcg, 6) if idcg > 0 else 0.0


def compute_mrr_gpt(grades):
    """
    MRR using binary GPT grades.
    MRR = 1 / rank_of_first_relevant_doc
    If no relevant doc → MRR = 0
    """
    for i, g in enumerate(grades):
        if g == 1:
            return round(1.0 / (i + 1), 6)
    return 0.0


def compute_srr(priv_labels):
    """SRR = privileged_count / underrepresented_count."""
    priv  = sum(1 for p in priv_labels if p == "Privileged")
    under = sum(1 for p in priv_labels if p == "Underrepresented")
    return round(priv / (under if under > 0 else 0.5), 6)


def compute_spd(priv_labels):
    """SPD = privileged_rate - 0.58  (corpus baseline)."""
    return round(
        sum(1 for p in priv_labels if p == "Privileged") / len(priv_labels)
        - CORPUS_BASELINE, 6
    )


# ══════════════════════════════════════════════════════════════════
# MAIN LOOP
# ══════════════════════════════════════════════════════════════════
print("="*62)
print("Computing P@10 + NDCG@10 + MRR + SRR + SPD …")
print(f"  Queries={len(queries)} | λ={LAMBDAS} | GPT calls={len(queries)*TOP_K_MMR*len(LAMBDAS)}")
print("="*62 + "\n")

rows, skipped = [], 0

for q in tqdm(queries, desc="Queries"):
    qid, qtext = q["query_id"], q["query"]

    try:
        nodes_50 = retriever_50.retrieve(qtext)
    except Exception as e:
        print(f"\n  [WARN] {qid}: {e}")
        skipped += 1
        continue
    if not nodes_50:
        skipped += 1
        continue

    cand_embs = st_model.encode(
        [n.get_content()[:500] for n in nodes_50],
        normalize_embeddings=True, show_progress_bar=False, batch_size=64)
    q_emb = st_model.encode(qtext, normalize_embeddings=True)

    all_privs = [
        get_privilege_label(n.node.metadata.get("primary_institution", ""))
        for n in nodes_50
    ]

    for lam in LAMBDAS:
        # MMR select top-10
        sel_idx   = mmr_rerank(q_emb, cand_embs, lam=lam, k=TOP_K_MMR)
        sel_nodes = [nodes_50[i] for i in sel_idx]
        sel_privs = [all_privs[i] for i in sel_idx]

        # P@10 via GPT — also gives us binary grades for NDCG + MRR
        precision, grades = compute_precision_at_k(qtext, sel_nodes)

        # NDCG@10 using GPT binary grades
        ndcg = compute_ndcg_at_k_gpt(grades)

        # MRR using GPT binary grades
        mrr  = compute_mrr_gpt(grades)

        # SRR + SPD from privilege labels
        srr = compute_srr(sel_privs)
        spd = compute_spd(sel_privs)

        priv_cnt = sum(1 for p in sel_privs if p == "Privileged")

        rows.append({
            "query_id"        : qid,
            "query_type"      : q.get("query_type", ""),
            "primary_cat"     : q.get("primary_cat", ""),
            "region"          : q.get("region", ""),
            "query_privilege" : q.get("privilege", ""),
            "lambda"          : lam,
            "n_candidates"    : len(nodes_50),
            # ── 5 metrics ──────────────────────────────────────
            "precision_at_10" : precision,   # P@10  (GPT)
            "ndcg_at_10"      : ndcg,        # NDCG@10 (GPT grades)
            "mrr"             : mrr,         # MRR (GPT grades)
            "srr"             : srr,         # privileged/underrep
            "spd"             : spd,         # priv_rate - 0.58
            # ── supporting ─────────────────────────────────────
            "privileged_count": priv_cnt,
            "privileged_pct"  : round(priv_cnt / TOP_K_MMR * 100, 1),
            "corpus_baseline" : CORPUS_BASELINE,
        })

print(f"\n {len(queries)-skipped}/{len(queries)} queries done  ({skipped} skipped)")

# ── Save CSV ───────────────────────────────────────────────────────
df = pd.DataFrame(rows)
df.drop(columns=[], errors="ignore").to_csv(
    os.path.join(OUTPUT_DIR, "mmr_metrics_full.csv"), index=False)
print(f" Saved mmr_metrics_full.csv  ({len(df)} rows)")

# ── Aggregate per lambda ───────────────────────────────────────────
agg = df.groupby("lambda")[["precision_at_10","ndcg_at_10","mrr","srr","spd","privileged_pct"]]\
        .mean().round(4).reset_index()
agg.to_csv(os.path.join(OUTPUT_DIR, "final_comparison_table.csv"), index=False)

# ── Print final table ──────────────────────────────────────────────
print("\n" + "="*72)
print("  FINAL RESULTS — all 5 metrics per lambda")
print("="*72)
print(f"  {'λ':<6} | {'P@10':>7} | {'NDCG@10':>8} | {'MRR':>7} | {'SRR':>7} | {'SPD':>7} | {'Priv%':>7}")
print("  " + "─"*65)
for _, r in agg.iterrows():
    print(f"  {str(r['lambda']):<6} | {r['precision_at_10']:>7.4f} | "
          f"{r['ndcg_at_10']:>8.4f} | {r['mrr']:>7.4f} | "
          f"{r['srr']:>7.4f} | {r['spd']:>7.4f} | {r['privileged_pct']:>6.1f}%")

print("""
  METRIC GUIDE:
  P@10    → % of top-10 docs relevant (GPT judge)
  NDCG@10 → ranking quality accounting for position (GPT grades)
  MRR     → reciprocal rank of first relevant doc (GPT grades)
  SRR     → privileged/underrepresented ratio in top-10
  SPD     → privileged_rate - 0.58  (fairness vs corpus)
""")

# ── Breakdown by category ──────────────────────────────────────────
print("Breakdown by primary_cat × lambda:")
agg2 = df.groupby(["primary_cat","lambda"])[["precision_at_10","ndcg_at_10","mrr","spd"]]\
         .mean().round(4).reset_index()
print(agg2.to_string(index=False))
agg2.to_csv(os.path.join(OUTPUT_DIR, "metrics_by_category.csv"), index=False)

print("\n" + "="*72)
print("  ALL DONE!")
print(f"  Output: {OUTPUT_DIR}")
print("  ├── mmr_metrics_full.csv         (400 rows: query × lambda)")
print("  ├── final_comparison_table.csv   (aggregated per lambda)")
print("  └── metrics_by_category.csv      (per CS category)")
print("="*72)
```
